This notebook creates a `Dataset` object to store data from a folder of FITS files and visualizes a sample spectrogram. After preprocessing the data to remove noise, the sample spectrogram is visualized again.

```python
# Import necessary modules
import radburst.utils.utils as utils
from radburst.utils.dataset import Dataset

# Define the path to the data folder
data_path = '/mnt/c/Users/camer/OneDrive/Documents/radburst/data/sample_data'

# Create a Dataset object which loads all data from the given path (defined in dataset.py)
data = Dataset(data_dir= data_path)
```

```python
# Now we can plot a spectrogram from the dataset
utils.plot_spectrogram(data[0])
```

```python
# Preprocess the dataset and plot again
data.preprocess(standardize_rows = True, remove_vertical_lines = True)

# Plot the same spectrogram again after preprocessing
utils.plot_spectrogram(data[0])
```
